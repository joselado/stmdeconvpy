```python
# add the library
import os
import sys
sys.path.append("../src")

# and import the necessary libraries
import numpy as np
from stmdeconvpy import deconvolve
from stmdeconvpy import profiles
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})

import matplotlib.pyplot as plt
```

# See how to deconvolve a superconducting profile

```python
xs = np.linspace(-4.0,4.0,1000)

# define filtering signal (the STM tip)
yf = profiles.step_tip(delta=1.0,lowest=0.1,highest=1.0)(xs)
yf = yf/np.mean(yf)


# define the real signal
y0 = profiles.vortex_state(delta=0.3)(xs)
y0 = profiles.dynes_superconductor(delta=0.3,gamma=0.01)(xs)
y0 = y0/np.mean(y0)

# compute the dIdV associated to that DOS
(xc,yc) = deconvolve.dos2dIdV(xs,y0,xs,yf)

yc = yc/np.mean(yc)


# deconvolve the signal, from dIdV 2 DOS
xn,ydc = deconvolve.dIdV2dos(xs,yc,xs,yf,n=300,mode="minimize")
#ydc = ydc/np.max(ydc)

# and do it again to compare
(xc2,yc2) = deconvolve.dos2dIdV(xs,ydc,xs,yf)
yc2 = yc2/np.mean(yc2)
```

```python
fig = plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Tip DOS")
plt.plot(xs,yf,label="Tip DOS",c="blue")
plt.xlim([np.min(xs),np.max(xs)])
plt.xlabel("Energy [a.u.]")
plt.ylabel("dI/dV or DOS [a.u.]")



plt.subplot(1,3,2)
plt.title("Measured dI/dV")
plt.plot(xc,yc,label="Real dIdV",c="lightgreen",linewidth=3)
plt.scatter(xc2,yc2,label="dIdV from deconv. DOS",c="red")
plt.xlim([np.min(xs),np.max(xs)])
plt.xlabel("Energy [a.u.]")
plt.ylabel("dI/dV or DOS [a.u.]")
plt.legend(fontsize=12,loc='upper center',ncol=1)
plt.ylim([0.,3.])


plt.subplot(1,3,3)
plt.title("Surface DOS")
plt.plot(xs,y0,label="Real DOS",c="lightgreen",linewidth=3)
plt.scatter(xn,ydc,label="Deconvolved DOS",c="red")
plt.xlim([np.min(xs),np.max(xs)])
plt.xlabel("Energy [a.u.]")
plt.ylabel("dI/dV or DOS [a.u.]")
plt.legend(fontsize=12,loc='upper center',ncol=1)
plt.ylim([0.,3.])

plt.tight_layout()
```

# See how to deconvolve a bound state in a vortex

```python
xs = np.linspace(-4.0,4.0,1000)

# define filtering signal (the STM tip)
yf = profiles.step_tip(delta=1.0,lowest=0.1,highest=1.0)(xs)
yf = yf/np.mean(yf)


# define the real signal
y0 = profiles.vortex_state(delta=0.3)(xs)
y0 = y0/np.mean(y0)

y0 = y0 + 0.2*np.random.random(len(y0))

# compute the dIdV associated to that DOS
(xc,yc) = deconvolve.dos2dIdV(xs,y0,xs,yf)

yc = yc/np.mean(yc)


# deconvolve the signal, from dIdV 2 DOS
xn,ydc = deconvolve.dIdV2dos(xs,yc,xs,yf,n=300,mode="minimize")
#ydc = ydc/np.max(ydc)

# and do it again to compare
(xc2,yc2) = deconvolve.dos2dIdV(xs,ydc,xs,yf)
yc2 = yc2/np.mean(yc2)
```

```python
fig = plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Tip DOS")
plt.plot(xs,yf,label="Tip DOS",c="blue")
plt.xlim([np.min(xs),np.max(xs)])
plt.xlabel("Energy [a.u.]")
plt.ylabel("dI/dV or DOS [a.u.]")



plt.subplot(1,3,2)
plt.title("Measured dI/dV")
plt.plot(xc,yc,label="Real dIdV",c="lightgreen",linewidth=3)
plt.scatter(xc2,yc2,label="dIdV from deconv. DOS",c="red")
plt.xlim([np.min(xs),np.max(xs)])
plt.xlabel("Energy [a.u.]")
plt.ylabel("dI/dV or DOS [a.u.]")
plt.legend(fontsize=12,loc='upper center',ncol=1)
plt.ylim([0.,3.])


plt.subplot(1,3,3)
plt.title("Surface DOS")
plt.plot(xs,y0,label="Real DOS",c="lightgreen",linewidth=3)
plt.scatter(xn,ydc,label="Deconvolved DOS",c="red")
plt.xlim([np.min(xs),np.max(xs)])
plt.xlabel("Energy [a.u.]")
plt.ylabel("dI/dV or DOS [a.u.]")
plt.legend(fontsize=12,loc='upper center',ncol=1)
plt.ylim([0.,3.])

plt.tight_layout()
```
