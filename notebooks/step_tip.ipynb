```python
# add the library
import os
import sys
sys.path.append("../src")

# and import the necessary libraries
import numpy as np
from stmdeconvpy import deconvolve
from stmdeconvpy import profiles
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20})

import matplotlib.pyplot as plt
```

# See how the step tip transforms a superconducting profile

```python
xs = np.linspace(-4.0,4.0,1000) # voltages

# define the tip
yf = profiles.step_tip(delta=1.0,lowest=0.1,highest=1.0)(xs)
yf = yf/np.max(yf)

# define the real signal
y0 = profiles.dynes_superconductor(delta=1.0,gamma=0.01)(xs)
y0 = y0/np.max(y0)

# compute the dIdV associated to that DOS
(xc,yc) = deconvolve.dos2dIdV(xs,y0,xs,yf)
yc = yc/np.max(yc)




# now plot the results


fig = plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Tip transmission")
plt.plot(xs,yf,label="Tip DOS",c="blue")
plt.xlim([np.min(xs),np.max(xs)])
plt.xlabel("Energy [a.u.]")
plt.ylabel("dI/dV or DOS [a.u.]")



plt.subplot(1,3,2)
plt.title("Measured dI/dV")
plt.plot(xs,yc,label="Measured dI/dV (convolution)",c="green")
plt.xlim([np.min(xs),np.max(xs)])
plt.xlabel("Energy [a.u.]")
plt.ylabel("dI/dV or DOS [a.u.]")


plt.subplot(1,3,3)
plt.title("Real DOS")
plt.plot(xs,y0,c="red")
plt.xlim([np.min(xs),np.max(xs)])
plt.xlabel("Energy [a.u.]")
plt.ylabel("dI/dV or DOS [a.u.]")


plt.tight_layout()
```

# See how the step tip transforms a bound state in a vortex

```python
xs = np.linspace(-4.0,4.0,1000) # voltages

# define the tip
yf = profiles.step_tip(delta=1.0,lowest=0.1,highest=1.0)(xs)
yf = yf/np.max(yf)

# define the real signal
y0 = profiles.vortex_state(delta=0.3)(xs)
y0 = y0/np.max(y0)

# compute the dIdV associated to that DOS
(xc,yc) = deconvolve.dos2dIdV(xs,y0,xs,yf)
yc = yc/np.max(yc)


# now plot the results


fig = plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.title("Tip transmission")
plt.plot(xs,yf,label="Tip DOS",c="blue")
plt.xlim([np.min(xs),np.max(xs)])
plt.xlabel("Energy [a.u.]")
plt.ylabel("dI/dV or DOS [a.u.]")



plt.subplot(1,3,2)
plt.title("Measured dI/dV")
plt.plot(xs,yc,label="Measured dI/dV (convolution)",c="green")
plt.xlim([np.min(xs),np.max(xs)])
plt.xlabel("Energy [a.u.]")
plt.ylabel("dI/dV or DOS [a.u.]")


plt.subplot(1,3,3)
plt.title("Real DOS")
plt.plot(xs,y0,c="red")
plt.xlim([np.min(xs),np.max(xs)])
plt.xlabel("Energy [a.u.]")
plt.ylabel("dI/dV or DOS [a.u.]")


plt.tight_layout()
```
